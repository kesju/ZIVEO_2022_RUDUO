```python

import numpy as np


rpeaks_from_signal = np.array([10, 20, 30, 40, 50, 60, 70, 80])
print('rpeaks_from_signal:',rpeaks_from_signal)

rpeaks_from_json = np.array([10,30,50,70])
rpeaks_from_json = np.array([10,30,50,70,90,130,150,170,210,230,250,270])
print('rpeaks_from_json:', rpeaks_from_json)
symbols_from_json = np.array(['N','S','N','V'])
symbols_from_json = np.array(['N','S','N','V','N','S','N','V','N','S','N','V'])
print("symbols_from_json:",symbols_from_json)


atr_sample = rpeaks_from_signal

atr_symbol = np.full(len(rpeaks_from_signal), 'N',  dtype=str)

 
           # 1
# Fiksuojame, kiek rankiniu būdu pridėta rpikų, jos ignorojamos
# Tai reikšmės faile rpeaks_from_json kurių nėra faile rpeaks_from_signal
ab = np.setdiff1d(rpeaks_from_json, rpeaks_from_signal)

 
            # 2
# Surandame masyve rpeaks_from_signal reikšmes rpeaks, kurios yra ir rpeaks_from_json
# Jei reikšmių faile rpeaks_from_json yra mažiau, negu faile rpeaks_from_signal, tai dalis panaikinta
aa, ind_a, ind_b = np.intersect1d(rpeaks_from_signal, rpeaks_from_json, assume_unique=True,return_indices=True )
if (len(aa) == len(rpeaks_from_signal)):
    print("\nVisi Neurokit rpikai yra ir json:", len(aa))
else:
    print("\nPanaikinta:", len(rpeaks_from_signal) - len(aa))

print("len(ind_a):", len(ind_a), "len(ind_b):", len(ind_b))

# Šioms reikšmėms anotacijas perrašome iš json
# for i in range (len(ind_a)):
#     atr_symbol[ind_a[i]] = symbols_from_json[ind_b[i]]
atr_symbol[ind_a] = symbols_from_json[ind_b]    

 
            # 3
# Fiksuojame, kiek rankiniu būdu panaikinta rpikų, nustatome kur, šioms reikšmėms anotacijos bus priskirtos 'U'
ba = np.setdiff1d(rpeaks_from_signal, rpeaks_from_json, assume_unique=True)
print("Anotuojant pridėta rpikų:", len(ab), ",jos ignoruojamos. Panaikinta:", len(ba), ",joms priskiriama anotacija 'U'")

# Šioms reikšmėms anotacijos  priskiriamos 'U'
flag = np.size(ba)
if flag:
    sorter = np.argsort(rpeaks_from_signal)
    idxs = sorter[np.searchsorted(rpeaks_from_signal, ba, sorter=sorter)]
    print("\nPanaikintos reikšmės:")
    print(f"i     index   rpeak")
    for i in range(len(idxs)):
        print(f"{i} {idxs[i]:>9} {ba[i]:>7}")
    atr_symbol[idxs] = 'U'
else:
    print('Tokių reikšmių nėra')
print('\natr_sample:',atr_sample)    
print('atr_symbol:',atr_symbol)    



# Bandymas atrinkti dalį masyvo reikšmių panaudojant kitą masyvą
print("\nBandymas atrinkti dalį masyvo reikšmių panaudojant kitą masyvą")
rpeaks_from_json = [10,30,50,70,90,130,150,170,210,230,250,270]
idx = [1,3,5] # indeksų sąrašas

# 1-as variantas: numpy array 
print("\n1-as variantas: numpy array")
print('rpeaks_from_json:',rpeaks_from_json)
print('idx:', idx)
rpeaks_from_json_array = np.array(rpeaks_from_json)
selected = rpeaks_from_json_array[idx]
print('selected:', selected)
print(list(selected))

# 2-as variantas: list 
# From <https://www.adamsmith.haus/python/answers/how-to-index-a-list-with-another-list-in-python> 
print("\n2-as variantas: list")
print('rpeaks_from_json:',rpeaks_from_json)
print('idx:', idx)
selected = [rpeaks_from_json[i] for i in idx]
print('selected:', selected)

# 3-as variantas: list 
# https://www.geeksforgeeks.org/python-accessing-all-elements-at-given-list-of-indexes/
# - pvz. linke neveikiantis, klaida, reikia map įvilkti į list
print("\n3-as variantas: list")
print('rpeaks_from_json:',rpeaks_from_json)
print('idx:', idx)
selected = list(map(rpeaks_from_json.__getitem__, idx))
print('selected:', selected)


# 4-as variantas: list - neveikiantis
print("\n4-as variantas: list, neveikiantis")
print('rpeaks_from_json:',rpeaks_from_json)
print('idx:', idx)
# selected = rpeaks_from_json[idx]
# print('selected:', selected)

# initializing lists
test_list = [9, 4, 5, 8, 10, 14]
index_list = [1, 3, 4]
 
# printing original lists
print("Original list : " + str(test_list))
print("Original index list : " + str(index_list))
 
# using map() + __getitem__ to
# elements from list
res_list = list(map(test_list.__getitem__, index_list))
 
# printing result
print("Resultant list : " + str(res_list))
```
